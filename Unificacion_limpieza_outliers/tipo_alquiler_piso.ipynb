```python
import pandas as pd
import unidecode  # <- para eliminar tildes

# Cargar el CSV limpio
ruta = r"C:\Users\pablo\proyecto\comparador\alquiler_unificado_con_coordenadas_limpio.csv"
df = pd.read_csv(ruta)

# Función para extraer tipo desde texto
def detectar_tipo(texto, tipo):
    texto = unidecode.unidecode(str(texto).lower())  # minúsculas sin tildes
    return tipo in texto

# Palabras clave a buscar
tipos = {
    "piso": "es_piso",
    "casa": "es_casa",
    "atico": "es_atico",
    "estudio": "es_estudio",
    "apartamento": "es_apartamento",
    "duplex": "es_duplex",
    "chalet": "es_chalet",
    "finca": "es_finca",
    "loft": "es_loft"
}

# Crear columnas
for palabra, columna in tipos.items():
    df[columna] = df["titulo"].apply(lambda x: detectar_tipo(x, palabra))

# Guardar archivo actualizado
ruta_salida = r"C:\Users\pablo\proyecto\comparador\alquiler_con_tipos_detectados.csv"
df.to_csv(ruta_salida, index=False)

print(f"✅ Archivo actualizado guardado en:\n{ruta_salida}")
```
