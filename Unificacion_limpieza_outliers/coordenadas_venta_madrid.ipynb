```python
import pandas as pd

# Cargar el archivo CSV
df_venta = pd.read_csv("C:/Users/pablo/proyecto/comparador/venta_madrid_con_coordenadas_.csv")

# Eliminar duplicados basados en 'titulo'
df_venta = df_venta.drop_duplicates(subset="titulo", keep="first")

# Eliminar registros con valores nulos en columnas clave
df_venta = df_venta.dropna(subset=["precio_num", "habitaciones_num", "metros_num"])

# Rellenar valores nulos en 'superficie_util_num' con la media de la columna
df_venta["superficie_util_num"] = df_venta["superficie_util_num"].fillna(df_venta["superficie_util_num"].mean())

# Normalizar valores de 'certificado_energetico' (por ejemplo, convertir todos a un formato similar)
df_venta["certificado_energetico"] = df_venta["certificado_energetico"].fillna("No disponible")

# Rellenar valores nulos en 'tipo_vivienda' con "Desconocido"
df_venta["tipo_vivienda"] = df_venta["tipo_vivienda"].fillna("Desconocido")

# Eliminar la columna 'imagen_destacada' si no es necesaria
df_venta = df_venta.drop(columns=["imagen_destacada"])

# Eliminar filas con valores nulos en 'lat' y 'lon' (si la ubicación es crítica)
df_venta = df_venta.dropna(subset=["lat", "lon"])

# Verificar el resultado de la limpieza
df_venta.head()




```
