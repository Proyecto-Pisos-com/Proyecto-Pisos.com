```python
import pandas as pd

# Cargar el CSV limpio
ruta = r"C:\Users\pablo\proyecto\comparador\venta_madrid.csv"
df = pd.read_csv(ruta)

# Normalizar el texto del título para búsqueda
df["titulo_lower"] = df["titulo"].str.lower()

# Palabras clave y sus columnas nuevas
tipos = ["piso", "casa", "atico", "estudio", "apartamento", "duplex", "chalet", "finca", "loft"]

# Crear columnas booleanas
for tipo in tipos:
    col_name = f"es_{tipo}"
    df[col_name] = df["titulo_lower"].str.contains(rf"\b{tipo}\b", na=False)

# Eliminar columna temporal
df.drop(columns=["titulo_lower"], inplace=True)

# Guardar archivo actualizado
ruta_salida = r"C:\Users\pablo\proyecto\comparador\venta_con_tipos.csv"
df.to_csv(ruta_salida, index=False)

print(f"✅ Archivo con columnas de tipo guardado en:\n{ruta_salida}")
```
